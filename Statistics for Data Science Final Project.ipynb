```python
a = boston[boston["CHAS"] == 0]["MEDV"]
a
```

```python
b = boston[boston["CHAS"] == 1]["MEDV"]
b
```

```python
sp.stats.ttest_ind(a,b,axis=0,equal_var=True)
```

Since p-value more than alpha value of 0.05, we failed to reject null hypothesis since there is NO statistical significance.

```python
boston["AGE"].value_counts()
```

```python
boston.loc[(boston["AGE"] <= 35),'age_group'] = '35 years and younger'
boston.loc[(boston["AGE"] > 35) & (boston["AGE"]<70),'age_group'] = 'between 35 and 70 years'
boston.loc[(boston["AGE"] >= 70),'age_group'] = '70 years and older'
```

```python
boston
```

```python
low = boston[boston["age_group"] == '35 years and younger']["MEDV"]
mid = boston[boston["age_group"] == 'between 35 and 70 years']["MEDV"]
high = boston[boston["age_group"] == '70 years and older']["MEDV"]
```

```python
f_stats, p_value = sp.stats.f_oneway(low,mid,high,axis=0)
```

```python
print("F-Statistic={0}, P-value={1}".format(f_stats,p_value))
```

Since p-value more than alpha value of 0.05, we failed to reject null hypothesis since there is NO statistical significance.

```python
pearson,p_value = sp.stats.pearsonr(boston["NOX"],boston["INDUS"])
```

```python
print("Pearson Coefficient value={0}, P-value={1}".format(pearson,p_value))
```

Since the p-value (Sig. (2-tailed) < 0.05, we reject the Null hypothesis and conclude that there exists a relationship between Nitric Oxide and non-retail business acres per town.
